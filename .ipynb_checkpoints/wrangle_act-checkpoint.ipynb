# Data Wrangling of WeRateDogs twitter archive data

WeRateDogs is a twitter account that rate people's dogs with a humorous content about the dog. We have been provided with there twitter acrhive and below we are going to wrangle this data in 3 steps of gathering,assesing and cleaning.

> ## Gather

Here, we will be gathering the data from three sources. Two files have already been provided, one is 'twitter_archive_enhanced.csv' and the other is 'image_predictions.tsv'. Third file needs to be created by querying the twitter API and storing the data of each tweet. 

```python
# Importing the required packages

import pandas as pd
import requests
import os
import tweepy
import json
import datetime as dt
import re
import matplotlib
import matplotlib.pyplot as plt
import warnings
```

```python
# Downloading file 'image-predictions.tsv' from the given html

url='https://d17h27t6h515a5.cloudfront.net/topher/2017/August/599fd2ad_image-predictions/image-predictions.tsv'
r = requests.get(url)
file_name=url.split('/')[-1]

# Creating the file

with open(file_name, mode = 'wb') as file:
    file.write(r.content)

# Creating an object for the file

image_predict =pd.read_csv("image-predictions.tsv",sep='\t',encoding = 'utf-8')
```

```python
# Creating an object for the second file
twitter_data = pd.read_csv('twitter-archive-enhanced.csv',encoding = 'utf-8')
```

```python

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)

api = tweepy.API(auth, 
                 parser = tweepy.parsers.JSONParser(),
                 wait_on_rate_limit = True,
                 wait_on_rate_limit_notify = True)
```

```python
# Querying the twitter API

twitter_list = []
error_list = []

# Getting the object for all the tweets

for tweet_id in twitter_data['tweet_id']:
    try:
        page = api.get_status(tweet_id, tweet_mode = 'extended')
        
        favorites = page['favorite_count']
        retweets = page['retweet_count'] 
        followers = page['user']['followers_count']
        favourites = page['user']['favourites_count']
        date_stamp = page['created_at']
        
        twitter_list.append({'tweet_id': int(tweet_id),
                        'favorites': int(favorites),
                        'retweets': int(retweets),
                        'user_followers': int(followers),
                        'user_favourites': int(favourites),
                        'date_time': pd.to_datetime(date_stamp)})
    
    # Catching the exceptions of TweepError
    except Exception as e:
        error_list.append(tweet_id)
```

```python
# Lengh of the result
print("The lengh of the result", len(twitter_list))
# tweet_id of the errors
print("The lengh of the errors", len(error_list))
```

```python
error_list_new=[]

# Traversing the error list and getting there objects
for i in error_list:
    try:
        favorites = page['favorite_count']
        retweets = page['retweet_count'] 
        followers = page['user']['followers_count']
        favourites = page['user']['favourites_count']
        date_stamp = page['created_at']
        
        twitter_list.append({'tweet_id': int(tweet_id),
                        'favorites': int(favorites),
                        'retweets': int(retweets),
                        'user_followers': int(followers),
                        'user_favourites': int(favourites),
                        'date_time': pd.to_datetime(date_stamp)})
    
    # Catching the exceptions of TweepError
    except Exception as e:
        print(str(tweet_id)+ " _ " + str(i))
        error_list_new.append(i)
```

```python
# Length of the result
print("The lengh of the result", len(twitter_list))
# tweet_id of the errors
print("The lengh of the error", len(error_list_new))
```

```python
# Storing the tweet data in a text file
df=pd.DataFrame(twitter_list,columns=['tweet_id','favorites','retweets','user_followers','user_favourites','date_time'])
df.to_csv('tweet_json.txt',encoding = 'utf-8',index=False)

# Creating object of the third file
tweet_json=pd.read_csv('tweet_json.txt',encoding = 'utf-8')
```

### Issues found out after assessment of data

Below are the issues I could capture by visual and programmatic assessment of data

#### Quality

> tweet_json
- date_time should be datetime and not a string.
- Duplicate records are present for tweet id '666020888022790149'.

> twitter_data
- timestamp should be datetime and not a string.
- Missing images in some tweets.
- Retweeted ratings are present.
- No ratings are present on two tweet_ids(835152434251116546,746906459439529985).
- Column 'text' is not displayed properly.
- Columns with retweet data is not required (retweeted_status_id,retweeted_status_user_id,retweeted_status_timestamp,'in_reply_to_status_id','in_reply_to_user_id')
- Dog names are not correct in some rows.
- Incorrect dog rating extracted where numerator of rating is in decimal.
- Tweet id '810984652412424000'does not have a rating.
- Incorrect dog rating where the 'text' has more than one occurence of #/#. 
    - Tweets with one dog name mentioned but more than one occurence of #/# has first #/# considered as the dog rating.
    - Tweets with more than one dog name mentioned along with more than one occurence of #/# has first #/# considered as the dog rating.The later #/# are not considerd - Please note this is not addressed.
    
> image_predict 
- Missing some rows(2075 rows are present compared to 2356 rows in twitter_data)

#### Tidiness

- Four columns of dog types(pupper,doggo,puppo,floofer) in twitter data table should be in one column.
- The three prediction and three prediction_confidence related columns in image_predict table should be merged in one prediction and one prediction_confidence column respectively.
- Image-predict and tweet_json should be part of twitter_data.

#### 9. twitter_data: Incorrect dog names

##### Define
> Correct the dog names extracted from 'text' column.

##### Code

```python
# Current dog names

pd.value_counts(twitter_data_clean['name'])
```

```python
# Changing incorrect dog names
new_dog_name = []

for text in twitter_data_clean['text']:
    # Start with 'This is ' and the fisrt letter of the name is uppercase
    if 'This is ' in str(text) and re.match(r'[A-Z].*', text.split()[text.split().index('This') + 2]):
        new_dog_name.append(text.split()[text.split().index('This') + 2].strip(',').strip('.'))
    # Start with 'Meet ' and the fisrt letter of the name is uppercase
    elif str(text).startswith('Meet ') and re.match(r'[A-Z].*', text.split()[1]):
        new_dog_name.append(text.split()[1].strip(',').strip('.'))
    # Start with 'Say hello to ' and the fisrt letter of the name is uppercase
    elif str(text).startswith('Say hello to ') and re.match(r'[A-Z].*', text.split()[3]):
        new_dog_name.append(text.split()[3].strip(',').strip('.'))
    # Start with 'Here we have ' and the fisrt letter of the name is uppercase
    elif str(text).startswith('Here we have ') and re.match(r'[A-Z].*', text.split()[3]):
        new_dog_name.append(text.split()[3].strip(',').strip('.'))
    # Contain 'named' and the fisrt letter of the name is uppercase
    elif 'named' in str(text) and re.match(r'[A-Z].*', text.split()[text.split().index('named') + 1]):
        new_dog_name.append(text.split()[text.split().index('named') + 1].strip(',').strip('.'))
     # Contain 'name is' and the fisrt letter of the name is uppercase
    elif 'name is' in str(text) and re.match(r'[A-Z].*', text.split()[text.split().index('name') + 2]):
        new_dog_name.append(text.split()[text.split().index('name') + 2].strip(',').strip('.'))
    # No name specified or other style
    else:
        new_dog_name.append('None')

# Save the result in a new column 'new_dog_name'
twitter_data_clean['new_dog_name'] = new_dog_name
```

##### Test

```python
twitter_data_clean['new_dog_name'].value_counts()
```

#### 2. Tidiness: Image Prediction column

##### Define
> Create one prediction column and one prediction_confidence column. These columns will have the values of the best image prediction and its respective prediction confidence. The 9 prediction columns (p1,p1_conf,p1_dog,p2,p2_conf,p2_dog,p3,p3_conf,p3_dog) will thus be merged into 2 columns only.

##### Code

```python
# Prediction columns cleaning

predicted_breed=[]
confidence=[]

def prediction_func(pred):
    if(pred['p1_dog']==True):
        predicted_breed.append(pred['p1'])
        confidence.append(pred['p1_conf'])
    elif(pred['p2_dog']==True):
        predicted_breed.append(pred['p2'])
        confidence.append(pred['p2_conf'])
    elif(pred['p3_dog']==True):
        predicted_breed.append(pred['p3'])
        confidence.append(pred['p3_conf'])
    else:
        predicted_breed.append(None)
        confidence.append(0)

image_predict_clean.apply(prediction_func,axis=1)
image_predict_clean['predicted_breed']=predicted_breed
image_predict_clean['prediction_confidence']=confidence

image_predict_clean
```

```python
# Dropping the columns not required
image_predict_clean.drop(['p1','p1_conf','p1_dog','p2','p2_conf','p2_dog','p3','p3_conf','p3_dog'],axis=1,inplace=True)
```
